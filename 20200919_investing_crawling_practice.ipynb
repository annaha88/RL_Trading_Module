## 인베스팅 실시간 선물 데이터 파일다운로드 없이 불러오기
- 한 셀 통합버전

```python
# 인베스팅 실시간 선물 데이터 파일다운로드 없이 불러오기
# beautifulsoup을 통한 html 파싱을 막아놓았다. => 따라서 selenium 으로 웹드라이버를 통해 WebElement 를 선택하는 방법을 사용하였다.
import numpy as np
import pandas as pd

from selenium import webdriver
import time

driver = webdriver.Chrome('./webdriver/chromedriver')
time.sleep(1)
driver.get('https://www.investing.com/commodities/real-time-futures')
# 선물 항목명 얻기(names)
elements = driver.find_elements_by_tag_name('a')
names = [i.text for i in elements[315:351]]  # 이름 얻어오는 슬라이싱 구간이 변할수도 있을 것 같다. 다른 날 다시 테스트해 볼것.

# 종가, high, low, time 얻기
elements = driver.find_elements_by_tag_name('td')
classes = [element.get_attribute('class') for element in elements]
numbers=[name for name in classes if name.startswith("pid")]
numberss =[]
for number in numbers:
    try:
        element2 = driver.find_element_by_class_name(number)
        numberss.append(element2.text)
    except:
        numberss.append('NaN')

price_last=[numberss[4*i] for i in range(36)]
price_high=[numberss[4*i+1] for i in range(36)]
price_low=[numberss[4*i+2] for i in range(36)]
price_time=[numberss[4*i+3] for i in range(36)]

df = pd.DataFrame({
      'Commodity': names
    , 'Last':price_last
    , 'High':price_high
    , 'Low':price_low
    , 'Time':price_time
})

df.to_csv('./data/commodity_real_time_test2.csv'
         , sep=','
          , encoding='utf-8'
         )

driver.close()
```
